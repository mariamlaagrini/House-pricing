```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("kc_house_data.csv")
data.head()
```

```python
data.describe()
```

```python
reg = LinearRegression()
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
# Assign the parameters of the GradientBoostingRegressor to the 'params' variable
from sklearn import ensemble
from sklearn.metrics import mean_squared_error

params = {'n_estimators': 400, 'max_depth': 5, 'min_samples_split': 2, 'learning_rate': 0.1, 'loss': 'squared_error'}

# Use the params variable to create and fit the GradientBoostingRegressor model
clf = ensemble.GradientBoostingRegressor(**params)
clf.fit(x_train, y_train)

# Now you can access the parameters using params['n_estimators']
t_sc = np.zeros((params['n_estimators']), dtype=np.float64)
# y_pred = reg.predict(x_test)  # This line is not needed here
for i, y_pred in enumerate(clf.staged_predict(x_test)):
    # t_sc[i] = clf.loss_(y_test, y_pred) # Instead of using clf.loss_, calculate the loss using mean_squared_error
    t_sc[i] = mean_squared_error(y_test, y_pred)
testsc = np.arange((params['n_estimators'])) + 1
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```
